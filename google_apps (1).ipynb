```python
#Google PLaystore
import numpy as np 
import pandas as pd
```

Read Excel File into DataFrame

```python
playstore = pd.read_excel("GooglePlaystore.xlsx")
print(playstore.loc[[10472]])
print(len(playstore))
#display(playstore.head(10))
```

Drop row in Dataframe where "Reviews" Column says "3.0M" instead of an integer value. 

```python
for idx, row  in playstore.iterrows():
    if row["Reviews"] == "3.0M":
        playstore = playstore.drop(labels=10472,axis=0)
playstore.index = range(len(playstore))
#print(playstore.loc[[10472]])
print(len(playstore))
display(playstore.head(10))
```

Drop any row where any column value == "Varies with Device"

```python
print(len(playstore))
a = "Varies with device"
#i=0
for index, row in playstore.iterrows():
    for x in row:
        if x == a:
            #i = i+1
            playstore = playstore.drop(index)
            break
playstore.index = range(len(playstore))        
#print(i)
print(len(playstore))
display(playstore.head(10))
```

Convert "Android Ver" column values to floating point types up to one decimal

```python
playstore['Android Ver1'] = playstore["Android Ver"].str[:3]
playstore=playstore.drop('Android Ver', axis = 1)
playstore = playstore.rename(columns ={'Android Ver1': 'Android Ver'})
playstore['Android Ver'] = playstore['Android Ver'].astype(float)
#for x in playstore['Android Ver']:
#    display(type(x))
#    break
display(playstore.head(10))
```

On the Installs Column get rid of '+' at the end, and the ',' so you can convert the column values to intger types

```python
playstore['Installs'] = playstore['Installs'].str.strip('+') 
playstore['Installs'] = playstore['Installs'].replace(',','', regex=True)
playstore['Installs'] = playstore['Installs'].astype(int)
#for x in playstore['Installs']:
#    display(type(x))
#    break
display(playstore.head(10))
```

Drop any row where the Rating column is nan, the Reviews column is <100 and Installs column is < 50000.

```python
print(len(playstore))
for index, row in playstore.iterrows():
    if np.isnan(row['Rating']) and row['Reviews'] < 100 and row['Installs'] < 50000:
            playstore = playstore.drop(index)
playstore= playstore.reset_index(drop=True)
#display(playstore)
playstore.iloc[[22]]
```

If a row's Rating column is nan find the average rating of other google apps in that same category. Then replace Rating nan value with the average value.

```python
for index, row in playstore.iterrows():
    if np.isnan(row['Rating']):
        #display(row['Rating'])
        #display(playstore.loc[index, 'Rating'])
        a = row['Category']
        #display(a)
        b = playstore.loc[playstore['Category'] == a]
        c= b['Rating'].mean()
        c = round(c,2)
        playstore.loc[index, 'Rating'] = c
        #display(playstore.loc[index, 'Rating'])
playstore.iloc[[22]]
```

Convert "Size" values to integer types. Orginal format was 7.0M or 7.0k.  

```python
playstore['Size']= playstore['Size'].replace({'k': '*1e3', 'M': '*1e6'}, regex=True).map(pd.eval).astype(int)
playstore.iloc[[22]]
```

Find the Discrete Statistics for each category of apps.

```python
df = playstore.groupby("Category")['Rating'].describe().reset_index()
display(df)
```

Display the top 3 free apps based on rating, Installs, and then Reviews

```python
df1 = playstore.loc[playstore['Type'] == 'Free']
df1 = df1.astype({'Reviews':'float'})
def top3(x):
    ff = df1.set_index('App').groupby('Category')[x].nlargest(3)
    ff = ff.to_frame()
    return ff
aa= top3('Rating')
bb= top3('Installs')
cd= top3('Reviews')
display("Top 3 Free Apps by Category: Criteria = Rating", aa)         
display("Top 3 Free Apps by Category: Criteria = Installs", bb)
display("Top 3 Free Apps by Category: Criteria = Reviews", cd)
```

Create a Pie chart indicating the number of apps present in the Google Playstore broken up by Genre

```python
# look I read about dataframe.explode and I didn't think it would work to make "Art & Design;Creativity' into "[‘Art & Design’, Creativity’]" I though str.split was better. 
#Then I used explode to make them into different rows so I could count better. My assumption was that One app could count as two genres. 
gg = playstore.copy()
gg['Genres']=gg["Genres"].str.split(';')
```

```python
import matplotlib.pyplot as plt
gg = gg.explode('Genres')
yy = gg['Genres'].value_counts()
#display(yy)
labels = yy.index
data = yy
fig = plt.figure(figsize =(40,20))
plt.pie(data, labels = labels)
plt.show()
```
